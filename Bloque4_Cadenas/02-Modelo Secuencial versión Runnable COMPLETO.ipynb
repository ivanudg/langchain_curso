```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnableMap

f = open('../OpenAI_key.txt')
api_key = f.read()
llm = ChatOpenAI(openai_api_key=api_key)
```

```python
prompt1 = ChatPromptTemplate.from_template(
    "Dame un resumen del rendimiento de este trabajador:\n{revision_rendimiento}"
)
chain_1 = prompt1 | llm
```

```python
prompt2 = ChatPromptTemplate.from_template(
    "Identifica las debilidades de este trabajador dentro de este resumen de la revisión:\n{resumen_revision}"
)
chain_2 = prompt2 | llm
```

```python
workflow = RunnableMap({
    "resumen_revision": chain_1,
}) | chain_2
```

```python
entrada = {
    "revision_rendimiento": "Este trabajador ha demostrado buenas capacidades técnicas, pero tiene dificultades con la puntualidad y la comunicación en equipo."
}
salida = workflow.invoke(entrada)
print(salida)
```

```python
prompt3 = ChatPromptTemplate.from_template(
    "Crea un plan de mejora para ayudar en estas debilidades:\n{debilidades}"
)
chain_3 = prompt3 | llm
```

```python
# Flujo completo encadenando los tres pasos
workflow = (
    RunnableMap({"resumen_revision": chain_1})
    | RunnableMap({"debilidades": chain_2})
    | chain_3
)
```

```python
entrada = {
    "revision_rendimiento": "Este trabajador ha demostrado buenas capacidades técnicas, pero tiene dificultades con la puntualidad y la comunicación en equipo."
}
salida = workflow.invoke(entrada)
print(salida)
```
